## Exploração Analítica

```python
#Filtrar tipos de colunas
#Exemplo abaixo foi filtrado colunas que são do tipo object e as que são diferentes do tipo object

colunas_categorias = base_dados.columns[base_dados.dtypes == object]
colunas_numericas = base_dados.columns[base_dados.dtypes != object]

colunas_categorias, colunas_numericas
```

```python
# Informando a quantidade de dados por % = normalize = True
base_dados['city'].value_counts(normalize=True) * 100
```

```python
#Mostrando a qtde de informações por porcentagem (animais, andares)
for Coluna in colunas_categorias:
    analise = base_dados[Coluna].value_counts(normalize=True)*100
    
    print(Coluna, '\n', analise, '\n')
```

```python
# Correção nos dados
# Ajustando o Andar

#loc - localizar / iloc - passar parametro para alteração na base de dados

# Alterando o andar 301 p/ 30

base_dados.loc[base_dados['floor'] == '301']

base_dados.iloc[2562,5] = 30
base_dados.loc[base_dados['floor'] == 30]
```

```python
# Alterando a coluna '-'
# apply - Comando q permite fazer operações dentro da coluna

# Utilizando o lambda para realizar operação dentro da coluna
# Será percorrido a base de dados e onde estiver o '-', será trocado por 0
# e também será reescrita a base com esta alteração que estamos fazendo
base_dados['floor'] = base_dados['floor'].apply(lambda Registro: 0 if Registro == '-' else Registro)
```

```python
#Verificar ultimos registros da base
base_dados.tail()
```

```python
# Colocando a coluna floor para numerico
# Reescrevendo a base com as alterações
base_dados['floor'] = pd.to_numeric(base_dados['floor'])
```

```python
base_dados.head()
```

```python
# Analise das colunas numéricas
# Grid com gráficos para analisar de uma forma completa as informações.

# Tamanho - Criar uma imagem com uma certa dimensão
Figura, Eixo = plt.subplots(figsize=(20, 30))

# Cor de fundo
cor_fundo = '#f5f5f5'
Figura.set_facecolor(cor_fundo)

# Definição de paleta de cores para utilização nos gráficos
paleta_cores = sms.color_palette('flare', len(colunas_numericas)*2)

# Título - cor tamanho e expessura da fonte
plt.suptitle('Análise das variavéis numéricas', fontsize=22, color='#404040', fontweight=600)

# Estrutura do grid (qtde linhas colunas e posição inicial do grid)
Linhas = 7 #Todas as linhas numéricas
Colunas = 2 #(Boxplot - Distplot)
Posicao = 1 # Posição inicial do grid

# Percorrer as colunas numéricas (Loop para plotar os gráficos)

for coluna in colunas_numericas:
    
    #Plot no Grid(Boxplot)
    plt.subplot(Linhas, Colunas, Posicao)
    
    #Título
    plt.title(f'{coluna}', loc='left', fontsize=14, fontweight=200)
    
    #Plot - data é a base de dados e o y será as colunas que estou fazendo o loop
    # showmeans exibe um triangulo que é a média.
    sms.boxplot(data=base_dados, y=coluna, showmeans=True, saturation=0.75,
               linewidth=1, color = paleta_cores[Posicao], width=0.25)
    
    # Mudar - No loop ao realizar o processo de plotar, ele pula para o próximo grid (posicao +=1)
    Posicao += 1
    
    #Plot no Grid(Distplot)
    plt.subplot(Linhas, Colunas, Posicao)
    
    #Título
    plt.title(f'{coluna}', loc='left', fontsize=14, fontweight=200)
    
    #Plot - Diferença que no distplot você passa diretamente o valor da coluna
    #Posicao -1 para não mudar de cor ao mudar de grid
    sms.distplot(base_dados[coluna], color= paleta_cores[Posicao-1])
    
    # Mudar
    Posicao += 1

# Ajuste de Grid
# top - ajuste para deixar uma distancia do topo do gráfico
# hspace = espaço entre os gráficos
plt.subplots_adjust(top=0.95, hspace=0.3)
```
